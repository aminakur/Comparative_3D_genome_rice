```python
#TADs, bw, genes
import coolbox
from coolbox.api import *
```

```python
#visualizing overlapping sets of TADs
#hitad+hicexplorer TADs at 2kb resolution chr1
cool = '/scratch/ak8725/az_mrg/az2kb.cool'
hicexp_tads = '/scratch/ak8725/az_mrg/hicFindTADs/hicFindTADs1_out/az_2kb_domains.bed'
hitad_tads = '/scratch/ak8725/az_mrg/hitad/az_hitad2kb_0lvl_6.txt'
tads80 = '/scratch/ak8725/az_mrg/TADs_2kb_two_tools_intersect_80.bed'
test_region = "chr02:4000000-4200000"

data_dir = "/scratch/ak8725/az_mrg/fitcons/"
bw1 = f"{data_dir}/H3K18ac_true.bw"
bw2 = f"{data_dir}/H3K27ac_true.bw"
bw3 = f"{data_dir}/H3K27me3_true.bw"
bw4 = f"{data_dir}/H3K4me3_true.bw"
bw5 = f"{data_dir}/methylation_true.bw"
bw6 = f"{data_dir}/PRO-seq_true.bw"

frame = XAxis() + \
    HiCMat(cool, style='window', depth_ratio=0.3) + \
    HiCMat(cool, style='window', depth_ratio=0.3) + Title('TADs') + \
    TADCoverage(hicexp_tads, border_only=True, border_style = 'solid', border_color = 'blue', alpha=1.0) + \
    TADCoverage(hitad_tads, border_only=True, border_style = 'solid', border_color = 'green', alpha=1.0) + \
    HiCMat(cool, style='window', depth_ratio=0.3) + Title('80% overlapping TADs') + \
    TADCoverage(tads80, border_only=True, border_style = 'solid', border_color = 'darkred', alpha=1.0) + \
    BigWig(bw1, alpha=0.6) + Color("orange") + \
    Title("H3K18ac") + \
    BigWig(bw2, alpha=0.6) + Color("deepskyblue") + \
    Title("H3K27ac") + \
    BigWig(bw3, alpha=0.6) + Color("teal") + \
    Title("H3K27me3") + \
    BigWig(bw4, alpha=0.6) + Color("navy") + \
    Title("H3K4me3") + \
    BigWig(bw5, alpha=0.6) + Color("tomato") + \
    Title("Methylation") + \
    BigWig(bw6, alpha=0.6) + Color("orchid") + \
    Title("PRO-seq")

frame.plot(test_region)
```
